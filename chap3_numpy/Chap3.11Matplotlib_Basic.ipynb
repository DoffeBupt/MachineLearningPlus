## MatplotLib基础

```python
import matplotlib as mpl # 一般不完全引入
import matplotlib.pyplot as plt # 一般用这个模块
```

```python
import numpy as np
x = np.linspace(1,10,100)
x
```

```python
y = np.sin(x)
y
```

```python
plt.plot(x,y) # 前两个参数为x轴，y轴，把一个对象存了起来,一般要用show来让图片画出来，在画之前一般会各种丰富这个图，最后画
# 感觉就像是plt是一个对象，调用它的各种方法进行搞事情，然后调用它的show
plt.show() # 仿佛显示改版了hhh
```

```python
cosy = np.cos(x)
siny = y.copy()
```

```python
plt.plot(x,siny)
plt.plot(x,cosy)
plt.show()
# 仿佛现在jupyter会自动执行那个plt.show()并且之前的会被清除
```

```python
plt.plot(x,siny)
plt.plot(x,cosy,color='red')# 可以指定颜色，也可以用16进制码进行指定
plt.show()
```

```python
plt.plot(x,siny,linestyle='-.')
plt.plot(x,cosy,color='red',linestyle='--')# 同样可以指定绘制的样式，默认-
plt.show()
```

```python
plt.plot(x,siny,linestyle='-.')
plt.plot(x,cosy,color='red',linestyle='--')
plt.xlim(-5,15) # x的范围
plt.ylim(-0.5,0) # y的
plt.show()
```

```python
plt.plot(x,siny,linestyle='-.')
plt.plot(x,cosy,color='red',linestyle='--') 
plt.axis([-5,15,-0.5,0]) # 前两个是x的，后俩是y的
plt.show()
```

```python
plt.plot(x,siny,linestyle='-.')
plt.plot(x,cosy,color='red',linestyle='--')
plt.xlabel('x-axis')# x的标签
plt.ylabel('y-value')# y的标签
plt.show()
```

```python
plt.plot(x,siny,linestyle='-.',label='sinx') # label代表对应的线的标签
plt.plot(x,cosy,color='red',linestyle='--',label='cosx') # 同上
plt.xlabel('x-axis')# x的标签
plt.ylabel('y-value')# y的标签
plt.legend()# 加上这句话，代表显示label里的内容
plt.show()
```

```python
plt.plot(x,siny,linestyle='-.',label='sinx')
plt.plot(x,cosy,color='red',linestyle='--',label='cosx')
plt.xlabel('x-axis')
plt.ylabel('y-value')
plt.legend()
plt.title('TITLE') # 为整个图片绘制一个大的标题
plt.show()
```

## 散点图 Scatter Plot

```python
plt.scatter(x,siny)
plt.show()
```

```python
plt.scatter(x,siny)
plt.scatter(x,cosy,color='red') # 散点图，多为了显示二维特征之间的关系
plt.show()
```

```python
x = np.random.normal(0,1,10000)
y = np.random.normal(0,1,10000)
plt.scatter(x,y,alpha = 0.15) # alpha不透明度
plt.show() # 可以看到一个很标准的二维正态分布图
```
